```python
import pandas as pd
import numpy as np
```

```python
x = pd.read_excel('/content/경륜 DB_230904_Rating 추가_v2.00.xlsx', sheet_name='현역선수명단')
```

```python
x.isnull().sum()
```

```python
x['아마경력'].fillna(0,inplace=True)
x['혈액형'].fillna('모름',inplace=True)
x['출신지'].fillna('무소속',inplace=True)
x['출신팀'].fillna('무소속',inplace=True)
x['대퇴(위좌)'].fillna(x['대퇴(위좌)'].mean(),inplace=True)
x['대퇴(위우)'].fillna(x['대퇴(위우)'].mean(),inplace=True)
x['결혼'].fillna('미혼',inplace=True)
x['고등학교'].fillna('무소속',inplace=True)
x['흡연유무'].fillna('모름',inplace=True)
```

```python
x = x.drop(columns=['기타사항','출신팀.1', '선수번호고유값'])
x = x.iloc[1:-1,:]
x
```

```python
x = x.set_index(keys='선수번호')
```

```python
x.info()
```

```python
x = x.drop(columns=['Unnamed: 23', '출생년도'])
```

```python
from tables.table import Column
x.rename(columns = {'Unnamed: 24':'나이'}, inplace=True)
```

```python
x = x.astype(({'하퇴(위좌)':'float'}))
x = x.astype(({'하퇴(위우)':'float'}))
x = x.astype(({'신장':'float'}))
x = x.astype(({'체중':'float'}))
x = x.astype(({'시력(좌)':'float'}))
x = x.astype(({'시력(우)':'float'}))
x = x.astype(({'기수':'int'}))
x = x.astype(({'가슴둘레':'float'}))
x = x.astype(({'나이':'int'}))
```

```python
x.아마경력.value_counts()
```

```python
x.아마경력 = x.아마경력.replace({
    '선수':1,
    '비선수':2,
    '적성자':3,
    '용인대':4,
    '동호인':5,
    '한국체육대':6
})
```

```python
x.혈액형 = x.혈액형.replace({
    'A':1,
    'B':2,
    'O':3,
    'AB':4,
    '모름':5
})
```

```python
x.결혼 = x.결혼.replace({
   '기혼':1,
   '미혼':0
})
```

```python
x.흡연유무.value_counts()
```

```python
x.흡연유무 = x.흡연유무.replace({
   '무':1,
   '유':0,
   'X':1,
   '흡연':0,
   '흠연':0,
   '모름':1,
   '-':1
})
```

```python
x = x.drop(columns=['이름'])
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x['출신지'] = le.fit_transform(x['출신지'])
le = LabelEncoder()
x['출신팀'] = le.fit_transform(x['출신팀'])
le = LabelEncoder()
x['고등학교'] = le.fit_transform(x['고등학교'])
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x.loc[:,:] = std.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=7)
km.fit(x)
```
